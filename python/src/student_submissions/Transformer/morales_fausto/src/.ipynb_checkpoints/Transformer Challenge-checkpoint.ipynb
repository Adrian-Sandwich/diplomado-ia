```python
#Importar librerías
import os
import matplotlib.pyplot as plt

os.environ["KERAS_BACKEND"] = "tensorflow"

import pathlib
import random
import string
import re
import numpy as np

import tensorflow.data as tf_data
import tensorflow.strings as tf_strings
from tensorflow.keras.layers import Embedding, TextVectorization
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

import keras
from keras import layers
from keras import ops
from keras.layers import TextVectorization
```

```python
#Descargar Data
text_file = keras.utils.get_file(
    fname="spa-eng.zip",
    origin="http://storage.googleapis.com/download.tensorflow.org/data/spa-eng.zip",
    extract=True,
)

text_file = pathlib.Path(text_file).parent / "spa-eng_extracted" / "spa-eng" / "spa.txt"

#Parceo de datos:
with open(text_file) as f:
    lines = f.read().split("\n")[:-1]
text_pairs = []
for line in lines:
    eng, spa = line.split("\t")
    spa = "[start] " + spa + " [end]"
    text_pairs.append((eng, spa))

#visualizar parceo:
for _ in range(5):
    print(random.choice(text_pairs))



#definición de data prueba-entrenamiento
random.shuffle(text_pairs)
num_val_samples = int(0.15 * len(text_pairs))
num_train_samples = len(text_pairs) - 2 * num_val_samples
train_pairs = text_pairs[:num_train_samples]
val_pairs = text_pairs[num_train_samples : num_train_samples + num_val_samples]
test_pairs = text_pairs[num_train_samples + num_val_samples :]

print(f"{len(text_pairs)} total pairs")
print(f"{len(train_pairs)} training pairs")
print(f"{len(val_pairs)} validation pairs")
print(f"{len(test_pairs)} test pairs")



# Preprocesar los textos y crear un tokenizer
input_texts = [pair[0] for pair in train_pairs]  # Texto en inglés
target_texts = [pair[1] for pair in train_pairs]  # Texto en español

# Inicializar el Tokenizer para el inglés y español
input_tokenizer = Tokenizer()
target_tokenizer = Tokenizer()

# Ajustar el Tokenizer sobre los textos
input_tokenizer.fit_on_texts(input_texts)
target_tokenizer.fit_on_texts(target_texts)

# Crear los índices de palabras (word_index)
input_word_index = input_tokenizer.word_index  # word_index para inglés
target_word_index = target_tokenizer.word_index  # word_index para español

# Vocabulario del modelo
input_vocab_size = len(input_word_index) + 1  # Incluyendo el token para padding (0)
target_vocab_size = len(target_word_index) + 1  # Incluyendo el token para padding (0)

print(f"Vocabulario inglés (input): {input_vocab_size}")
print(f"Vocabulario español (target): {target_vocab_size}")



# Convertir las oraciones de entrada (inglés) a secuencias de índices
input_sequences = input_tokenizer.texts_to_sequences(input_texts)

# Convertir las oraciones de destino (español) a secuencias de índices
target_sequences = target_tokenizer.texts_to_sequences(target_texts)

# Asegurarnos de que todas las secuencias tengan la misma longitud
max_input_length = max([len(seq) for seq in input_sequences])
max_target_length = max([len(seq) for seq in target_sequences])

# Padding para que todas las secuencias tengan la misma longitud
encoder_input_data = pad_sequences(input_sequences, maxlen=max_input_length, padding='post')
decoder_input_data = pad_sequences(target_sequences, maxlen=max_target_length, padding='post')

# El target output (español) será la secuencia de palabras desplazada
# La palabra siguiente en la secuencia de español
decoder_output_data = pad_sequences(target_sequences, maxlen=max_target_length, padding='post')

print(f"Secuencias de entrada (inglés): {encoder_input_data.shape}")
print(f"Secuencias de salida (español): {decoder_input_data.shape}")

class CustomEmbeddingLayer(layers.Layer):
    def __init__(self, embedding_matrix, **kwargs):
        super().__init__(**kwargs)
        self.embedding_matrix = embedding_matrix
        self.embedding = layers.Embedding(
            input_dim=embedding_matrix.shape[0],
            output_dim=embedding_matrix.shape[1],
            weights=[embedding_matrix],
            trainable=False,  # No entrenar los embeddings
        )

    def call(self, inputs):
        return self.embedding(inputs)




# Capa personalizada para MultiHeadAttention que devuelve las puntuaciones de atención
class CustomMultiHeadAttention(layers.Layer):
    def __init__(self, num_heads, key_dim, **kwargs):
        super().__init__(**kwargs)
        self.num_heads = num_heads
        self.key_dim = key_dim
        self.attention_layer = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=key_dim, use_bias=False
        )

    def call(self, query, value, key, attention_mask=None):
        # Realizar la atención multi-cabeza
        attention_output, attention_weights = self.attention_layer(
            query=query,
            value=value,
            key=key,
            attention_mask=attention_mask,
            return_attention_scores=True,
        )
        return attention_output, attention_weights

# Función para cargar los embeddings GloVe
def load_glove_embeddings(file_path, embedding_dim):
    word2vec = {}
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            word2vec[word] = vector
    return word2vec

# Función para crear la matriz de embeddings
def create_embedding_matrix(word_index, word2vec, embedding_dim):
    num_words = len(word_index) + 1  # +1 para incluir el índice 0 para padding
    embedding_matrix = np.zeros((num_words, embedding_dim))
    
    for word, index in word_index.items():
        if word in word2vec:
            embedding_matrix[index] = word2vec[word]
    
    return embedding_matrix

# Cargar los embeddings de GloVe
glove_file_path = './glove.6B.100d.txt'  # Ruta del archivo GloVe
embedding_dim = 100  # Elige el tamaño adecuado según tu archivo GloVe
word2vec = load_glove_embeddings(glove_file_path, embedding_dim)

# Crear la matriz de embeddings a partir del word_index generado
embedding_matrix = create_embedding_matrix(input_word_index, word2vec, embedding_dim)

# Usar la matriz de embeddings para la capa de Embedding personalizada
embedding_layer = CustomEmbeddingLayer(embedding_matrix)



class TransformerEncoder(layers.Layer):
    def __init__(self, embed_dim, dense_dim, num_heads, **kwargs):
        super().__init__(**kwargs)
        self.embed_dim = embed_dim
        self.dense_dim = dense_dim
        self.num_heads = num_heads
        self.attention = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=embed_dim
        )
        self.dense_proj = keras.Sequential(
            [
                layers.Dense(dense_dim, activation="relu"),
                layers.Dense(embed_dim),
            ]
        )
        self.layernorm_1 = layers.LayerNormalization()
        self.layernorm_2 = layers.LayerNormalization()
        self.supports_masking = True

    def call(self, inputs, mask=None):
        if mask is not None:
            padding_mask = ops.cast(mask[:, None, :], dtype="int32")
        else:
            padding_mask = None

        attention_output = self.attention(
            query=inputs, value=inputs, key=inputs, attention_mask=padding_mask
        )
        proj_input = self.layernorm_1(inputs + attention_output)
        proj_output = self.dense_proj(proj_input)
        return self.layernorm_2(proj_input + proj_output)

    def get_config(self):
        config = super().get_config()
        config.update(
            {
                "embed_dim": self.embed_dim,
                "dense_dim": self.dense_dim,
                "num_heads": self.num_heads,
            }
        )
        return config

def positional_encoding(length, depth):
  depth = depth/2

  positions = np.arange(length)[:, np.newaxis]     # (seq, 1)
  depths = np.arange(depth)[np.newaxis, :]/depth   # (1, depth)
  
  angle_rates = 1 / (10000**depths)         # (1, depth)
  angle_rads = positions * angle_rates      # (pos, depth)

  pos_encoding = np.concatenate(
      [np.sin(angle_rads), np.cos(angle_rads)],
      axis=-1) 

  return tf.cast(pos_encoding, dtype=tf.float32)

class PositionalEmbedding(tf.keras.layers.Layer):
    def __init__(self, vocab_size, d_model):
        super().__init__()
        self.d_model = d_model
        self.embedding = tf.keras.layers.Embedding(vocab_size, d_model, mask_zero=True)
        self.pos_encoding = positional_encoding(length=2048, depth=d_model)

    def compute_mask(self, *args, **kwargs):
        return self.embedding.compute_mask(*args, **kwargs)

    def call(self, x):
        length = tf.shape(x)[1]  # Esto asegura que la longitud de la secuencia es dinámica
        x = self.embedding(x)  # Embedding de palabras
        # Escalar y añadir las posiciones
        x *= tf.math.sqrt(tf.cast(self.d_model, tf.float32))
        x = x + self.pos_encoding[tf.newaxis, :length, :]  # Añadir codificación posicional
        return x




class TransformerDecoder(layers.Layer):
    def __init__(self, embed_dim, latent_dim, num_heads, **kwargs):
        super().__init__(**kwargs)
        self.embed_dim = embed_dim
        self.latent_dim = latent_dim
        self.num_heads = num_heads
        self.attention_1 = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=embed_dim
        )
        self.attention_2 = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=embed_dim
        )
        self.dense_proj = keras.Sequential(
            [
                layers.Dense(latent_dim, activation="relu"),
                layers.Dense(embed_dim),
            ]
        )
        self.layernorm_1 = layers.LayerNormalization()
        self.layernorm_2 = layers.LayerNormalization()
        self.layernorm_3 = layers.LayerNormalization()
        self.supports_masking = True

    def call(self, inputs, mask=None):
        inputs, encoder_outputs = inputs
        causal_mask = self.get_causal_attention_mask(inputs)

        if mask is None:
            inputs_padding_mask, encoder_outputs_padding_mask = None, None
        else:
            inputs_padding_mask, encoder_outputs_padding_mask = mask

        attention_output_1 = self.attention_1(
            query=inputs,
            value=inputs,
            key=inputs,
            attention_mask=causal_mask,
            query_mask=inputs_padding_mask,
        )
        out_1 = self.layernorm_1(inputs + attention_output_1)

        attention_output_2 = self.attention_2(
            query=out_1,
            value=encoder_outputs,
            key=encoder_outputs,
            query_mask=inputs_padding_mask,
            key_mask=encoder_outputs_padding_mask,
        )
        out_2 = self.layernorm_2(out_1 + attention_output_2)

        proj_output = self.dense_proj(out_2)
        return self.layernorm_3(out_2 + proj_output)

    def get_causal_attention_mask(self, inputs):
        input_shape = ops.shape(inputs)
        batch_size, sequence_length = input_shape[0], input_shape[1]
        i = ops.arange(sequence_length)[:, None]
        j = ops.arange(sequence_length)
        mask = ops.cast(i >= j, dtype="int32")
        mask = ops.reshape(mask, (1, input_shape[1], input_shape[1]))
        mult = ops.concatenate(
            [ops.expand_dims(batch_size, -1), ops.convert_to_tensor([1, 1])],
            axis=0,
        )
        return ops.tile(mask, mult)

    def get_config(self):
        config = super().get_config()
        config.update(
            {
                "embed_dim": self.embed_dim,
                "latent_dim": self.latent_dim,
                "num_heads": self.num_heads,
            }
        )
        return config
# Hiperparámetros
embed_dim = 256
latent_dim = 2048
num_heads = 8
vocab_size = 10000  # Ajusta según tu tamaño de vocabulario
d_model = embed_dim


# Primero, aplica la capa PositionalEmbedding sobre los inputs
encoder_inputs = keras.Input(shape=(None,), dtype="int64", name="encoder_inputs")
x = PositionalEmbedding(vocab_size=vocab_size, d_model=d_model)(encoder_inputs)

# Luego, pasa el resultado al TransformerEncoder
encoder_outputs = TransformerEncoder(embed_dim, latent_dim, num_heads)(x)
encoder = keras.Model(encoder_inputs, encoder_outputs)

# Definir las entradas para el decodificador
decoder_inputs = keras.Input(shape=(None,), dtype="int64", name="decoder_inputs")
encoded_seq_inputs = keras.Input(shape=(None, embed_dim), name="decoder_state_inputs")

# Asegúrate de que la salida de la capa de PositionalEmbedding tenga las dimensiones correctas.
x = PositionalEmbedding(vocab_size=vocab_size, d_model=d_model)(encoder_inputs)
print(f"Shape after PositionalEmbedding: {x.shape}")

# Pasar las entradas correctamente al TransformerDecoder
x = TransformerDecoder(embed_dim, latent_dim, num_heads)([x, encoder_outputs])

# Agregar dropout y capa final de softmax
x = layers.Dropout(0.5)(x)
decoder_outputs = layers.Dense(vocab_size, activation="softmax")(x)

# Crear el modelo final del decodificador
decoder = keras.Model([decoder_inputs, encoded_seq_inputs], decoder_outputs)

# Crear el modelo completo
transformer = keras.Model(
    {"encoder_inputs": encoder_inputs, "decoder_inputs": decoder_inputs},
    decoder_outputs,
    name="transformer",
)

```

```python
#Entrenamiento
epochs = 40  # This should be at least 30 for convergence

transformer.summary()
transformer.compile(
    "rmsprop",
    loss=keras.losses.SparseCategoricalCrossentropy(ignore_class=0),
    metrics=["accuracy"],
)
history=transformer.fit(train_ds, epochs=epochs, validation_data=val_ds)
```

```python
#Traduciendo:
spa_vocab = spa_vectorization.get_vocabulary()
spa_index_lookup = dict(zip(range(len(spa_vocab)), spa_vocab))
max_decoded_sentence_length = 20


def decode_sequence(input_sentence):
    tokenized_input_sentence = eng_vectorization([input_sentence])
    decoded_sentence = "[start]"
    for i in range(max_decoded_sentence_length):
        tokenized_target_sentence = spa_vectorization([decoded_sentence])[:, :-1]
        predictions = transformer(
            {
                "encoder_inputs": tokenized_input_sentence,
                "decoder_inputs": tokenized_target_sentence,
            }
        )

        # ops.argmax(predictions[0, i, :]) is not a concrete value for jax here
        sampled_token_index = ops.convert_to_numpy(
            ops.argmax(predictions[0, i, :])
        ).item(0)
        sampled_token = spa_index_lookup[sampled_token_index]
        decoded_sentence += " " + sampled_token

        if sampled_token == "[end]":
            break
    return decoded_sentence


test_eng_texts = [pair[0] for pair in test_pairs]
for _ in range(5):
    input_sentence = random.choice(test_eng_texts)
    translated = decode_sequence(input_sentence)
    print(input_sentence,";",translated)
```
