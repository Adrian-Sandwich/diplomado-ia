###Convert the book into text (.txt) file

````python
## Function to convert from pdf to txt based on the code from https://www.geeksforgeeks.org/convert-pdf-to-txt-file-using-python/

def pdf_to_text(pdf_path, output_txt):
    txt = ''
    # Open the PDF file in read-binary mode
    with open(pdf_path, 'rb') as pdf_file:
        # Create a PdfReader object instead of PdfFileReader
        pdf_reader = PdfReader(pdf_file)

        # Initialize an empty string to store the text
        text = ''


        for page_num in range(len(pdf_reader.pages)):
            if page_num>= 4: ## Book text starts on page 5
              page = pdf_reader.pages[page_num]
              text += page.extract_text()
            else:
                text = ''

    # Write the extracted text to a text file
    with open(output_txt, 'w', encoding='utf-8') as txt_file:
        txt_file.write(text)

if __name__ == "__main__":
    pdf_path = book_path

    output_txt = path + 'LesMiserables_ff.txt'

    pdf_to_text(pdf_path, output_txt)

    print("PDF converted to text successfully!")
````

### Standardize the text to not include upper characters, punctuation and accents

````python
## Standardize (no upper chars, no whitespace, no punctuation, accents, if possible)

text = txt
text_re = re.sub(r"[\[\]]", "",text)
text_re = text_re.lower()
text_re = re.sub(r'[^\w\s]', '', text_re)
text_re = re.sub(r"[!?-]", "", text_re)
text_re = text_re.replace('\n', '')
````

````python
text_re
````

````python
def remove_accented_char(texto):
    # Normalizar el texto a la forma NFD
    texto = unicodedata.normalize("NFD", texto)

    # Reemplazar los caracteres diacríticos, pero dejando la "ñ" intacta
    texto = re.sub(r"(?<!n)[\u0300-\u036f]", "", texto)

    # Volver a la forma NFC para evitar problemas de codificación
    return unicodedata.normalize("NFC", texto)
````

````python
text_re = remove_accented_char(text_re)
````

### Define Counter class to get all the information needed from the text

````python
## Python has a module named collections and a method called Counter you can find them here  https://docs.python.org/3/library/collections.html
## We will use the same idea to create our version of the Counter

class CounterWords:
    def __init__(self, iterable=None):
        """Initialize the counter with an iterable as optional , in this case the iterable is the text"""
        self.datos = {}
        if iterable:
            self.update(iterable)

    def update(self, iterable):
        """ Update the counter with the words find on the iterable """
        if isinstance(iterable, str):
            iterable = re.findall(r'\b\w+\b', iterable.lower())  # Divide word by word ignoring punctuation

        for word in iterable:
            self.datos[word] = self.datos.get(word, 0) + 1 # if the word exists then add 1, if not initialize counter on 1


    def __getitem__(self, key):
        """ Access to the elements of the dictionary """
        return self.datos.get(key, 0)

    def most_common(self, n=None, desc= True):
        """ Show the nth most frecuent words """
        ordenado = sorted(self.datos.items(), key=lambda x: x[1], reverse=desc)
        return ordenado if n is None else ordenado[:n]

    def original_len(self, iterable):
          """ Show the lenght, word cnt of the original file """
          if isinstance(iterable, str):
            iterable = re.findall(r'\b\w+\b', iterable.lower())

          return  len(iterable)

    def unique_len(self, iterable):
          """ Show the lenght of unique, unique word cnt of the original file """
          if isinstance(iterable, str):
            iterable = re.findall(r'\b\w+\b', iterable.lower())

          return  len(set(iterable))


    def save_to_parquet(self, filename="vocabulary.parquet"):
        """ Save the dictionary on parquet file """
        df = pd.DataFrame(self.datos.items(), columns=["word", "frequency"])
        df.to_parquet(filename, engine="pyarrow", index=False)
        print(f"File Saved on:  {filename}")


    def __repr__(self):

        return f"CounterWords({self.datos})"
````
