```python
import fitz  # PyMuPDF
import re
from unidecode import unidecode
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import pandas as pd
```

```python
def pdf_a_texto(ruta_pdf):
    doc = fitz.open(ruta_pdf)
    texto = ""
    for pagina in doc:
        texto += pagina.get_text("text") + "\n"
    return texto
```

```python
ruta_pdf = "/content/drive/My Drive/Los-miserables.pdf"
```

```python
texto = pdf_a_texto(ruta_pdf)
print("üìå Texto original (primeros 1000 caracteres):\n", texto[:1000])
```

```python
def limpiar_texto(texto):
    texto = texto.lower()  # Convertir a min√∫sculas
    texto = unidecode(texto)  # Eliminar acentos
    texto = re.sub(r'[^\w\s]', '', texto)  # Quitar puntuaci√≥n
    texto = re.sub(r'\d+', '', texto)  # Eliminar n√∫meros
    texto = re.sub(r'\s+', ' ', texto).strip()  # Eliminar espacios extra
    return texto

texto_limpio = limpiar_texto(texto)
print("üìå Texto limpio (primeros 1000 caracteres):\n", texto_limpio[:1000])
```

```python
# Tokenizar y eliminar stopwords
stop_words = set(stopwords.words('spanish'))  # Cargar stopwords en espa√±ol
```

```python
# Tokenizar
tokens = word_tokenize(texto_limpio)
```

```python
# Quitar stopwords
tokens_filtrados = [word for word in tokens if word not in stop_words]
```

```python
# Contar frecuencia de palabras
frecuencia_palabras = Counter(tokens_filtrados)
```

```python
# Obtener las 100 palabras m√°s y menos frecuentes
mas_frecuentes = frecuencia_palabras.most_common(100)
menos_frecuentes = frecuencia_palabras.most_common()[:-101:-1]
```

```python
print(f"üìå Total de palabras despu√©s de quitar stopwords: {len(tokens_filtrados)}")
```

```python
print(f"üìå Palabras √∫nicas despu√©s de quitar stopwords: {len(frecuencia_palabras)}")
```

```python
# Mostrar estad√≠sticas del vocabulario
print("üîπ 10 palabras m√°s frecuentes (sin stopwords):")
for palabra, freq in mas_frecuentes[:10]:
    print(f"{palabra}: {freq}")
```

```python
print("\nüîπ 10 palabras menos frecuentes (sin stopwords):")
for palabra, freq in menos_frecuentes[:10]:
    print(f"{palabra}: {freq}")
```

```python
# Generar estad√≠sticas del vocabulario
total_palabras = len(tokens_filtrados)
palabras_unicas = len(frecuencia_palabras)
longitud_promedio = sum(len(p) for p in tokens_filtrados) / total_palabras
```

```python
print(f'üìå N√∫mero total de palabras (sin stopwords): {total_palabras}')
print(f'üìå N√∫mero de palabras √∫nicas (sin stopwords): {palabras_unicas}')
print(f'üìå Longitud promedio de las palabras: {longitud_promedio:.2f}')
```
